```python
df.groupby('Gender').size().plot(kind = 'pie', autopct="%.2f",
                                title = 'Gender Ratio', figsize = (5,3))
                            
```

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize = (5,3))
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct="%.2f",
                                title = 'Purchase', figsize = (5,3))
```

```python
df.groupby('Gender').mean()['Purchase'].plot(kind = 'pie', autopct="%.2f",
                                title = 'Purchase Average', figsize = (5,3))
```

```python
ssn.set(rc={'figure.figsize': (12,6)})
ssn.countplot(x= 'Age' , hue='Gender', data= df )
```
